```python
import matplotlib.pyplot as plot
import numpy as np
```

```python
k=2

seet = [[55, 150, 100, 2, 1350],
[55, 200, 120, 8.3, 3050],
[55,	220,	130,	8.3,	6780],
[65,	656,	150,	8.3,	8475],
[65,	300,	220,	33,	18000]]

targ = [65, 656, 220, 33, 18000]

mulps = [65,
656,
220,
33,
18000]

omeg = np.empty(5)

for i in range(5):
    omeg[i] = 0.2/mulps[i]

print(omeg)
```

```python
d = np.empty(5)

# omeg = [0.003077, 0.003049, 0.0009091, 0.0060606, 0.0000111]

for i in range(5):
    d[i] = 0

for i in range(5):
    for j in range(5):
        a = abs(seet[i][j]-targ[j])
        b = a**k
        c = omeg[j]*b
        f = c**(1/k)
        d[i] += f

print(d)
```
