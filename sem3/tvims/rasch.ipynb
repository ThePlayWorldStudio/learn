```python
mu    = np.mean(data)
sigma = np.std(data, ddof=1)

x_sorted = np.sort(data)
n = len(data)
F_emp = np.arange(1, n+1) / n

def norm_cdf_approx(z):
    t = 1 / (1 + 0.2316419 * np.abs(z))
    poly = (0.319381530 * t - 0.356563782 * t**2 + 1.781477937 * t**3 -
            1.821255978 * t**4 + 1.330274429 * t**5)
    phi = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * z**2)
    cdf = 1 - phi * poly
    return np.where(z >= 0, cdf, 1 - cdf)

z_sorted = (x_sorted - mu) / sigma
F_theor_points = norm_cdf_approx(z_sorted)

D_plus  = np.max(F_emp - F_theor_points)
D_minus = np.max(F_theor_points - np.concatenate(([0.], F_emp[:-1])))
D = max(D_plus, D_minus)

diff = np.abs(F_emp - F_theor_points)
idx_max = np.argmax(diff)
x_max_dev = x_sorted[idx_max]
F_emp_at_max = F_emp[idx_max]
F_theor_at_max = F_theor_points[idx_max]

x_grid = np.linspace(x_sorted.min() - 2, x_sorted.max() + 2, 500)
z_grid = (x_grid - mu) / sigma
F_theor = norm_cdf_approx(z_grid)


plt.figure(figsize=(11, 6))

plt.step(x_sorted, F_emp, where='post', color='navy', lw=2.3,
         label='Эмпирическая Fₙ(x)')

plt.plot(x_grid, F_theor, color='darkred', lw=2.5, ls='--',
         label=f'Нормальная N({mu:.2f}, {sigma:.2f}²)')

plt.plot(x_sorted, F_emp, 'o', ms=5, color='navy', alpha=0.6)

plt.plot([x_max_dev, x_max_dev], [F_emp_at_max, F_theor_at_max],
         color='limegreen', lw=3.2, alpha=0.9, zorder=10)
plt.plot(x_max_dev, F_emp_at_max, 'o', ms=10, color='limegreen', mec='black', mew=1.5,
         label=f'Макс. отклонение Z = {D:.4f}')
plt.plot(x_max_dev, F_theor_at_max, 'o', ms=10, color='limegreen', mec='black', mew=1.5)

plt.title('Эмпирическая и Теоретическая функции', fontsize=13)
plt.xlabel('x', fontsize=12)
plt.ylabel('F(x)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(loc='lower right', fontsize=10)
plt.ylim(-0.05, 1.05)
plt.xlim(-16, 7)

plt.tight_layout()
plt.show()
```
